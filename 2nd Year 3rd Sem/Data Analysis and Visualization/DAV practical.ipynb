```python
import numpy as np
# create a 1D array with some zero, non-zero and NaN elements
ARR3 = np.array([0, 1, 2, np.nan, 3, 0, 4, np.nan, 5])
print("ARR3:")
print(ARR3)
# test whether the elements are zero, non-zero and NaN
zero = np.equal(ARR3, 0)
nonzero = np.not_equal(ARR3, 0)
nan = np.isnan(ARR3)
print("Zero elements:", zero)
print("Non-zero elements:", nonzero)
print("NaN elements:", nan)
# record the indices of these elements in three separate arrays
zero_indices = np.where(zero)[0]
nonzero_indices = np.where(nonzero)[0]
nan_indices = np.where(nan)[0]
print("Indices of zero elements:", zero_indices)
print("Indices of non-zero elements:", nonzero_indices)
print("Indices of NaN elements:", nan_indices)
```

Do the following using PANDAS Series:

a. Create a series with 5 elements. Display the series sorted on index and also sorted on values seperately

b. Create a series with N elements with some duplicate values. Find the minimum and maximum ranks 
assigned to the values using ‘first’ and ‘max’ methods

c. Display the index value of the minimum and maximum element of a Series

```python
import pandas as pd
# create a series with 5 elements
s1 = pd.Series([5, 2, 4, 1, 3], index=['a', 'b', 'd', 'e', 'c'])
print("s1:")
print(s1)
# display the series sorted on index
s1_sorted_index = s1.sort_index()
print("s1 sorted on index:")
print(s1_sorted_index)
# display the series sorted on values
s1_sorted_values = s1.sort_values()
print("s1 sorted on values:")
print(s1_sorted_values)
```

```python
# create a series with N elements with some duplicate values
N = 10
s2 = pd.Series(np.random.randint(0, 5, size=N)) # change the range as per your requirement
print("s2:")
print(s2)
# find the minimum and maximum ranks assigned to the values using 'first' and 'max' methods
s2_rank_first = s2.rank(method='first')
s2_rank_max = s2.rank(method='max')
print("Minimum rank assigned to the values using 'first' method:", s2_rank_first.min())
print("Maximum rank assigned to the values using 'first' method:", s2_rank_first.max())
print("Minimum rank assigned to the values using 'max' method:", s2_rank_max.min())
print("Maximum rank assigned to the values using 'max' method:", s2_rank_max.max())
```

```python
# create a series with some elements
s3 = pd.Series([7, 2, 9, 4, 6, 3, 8, 5, 1])
print("s3:")
print(s3)
# display the index value of the minimum and maximum element of the series
s3_min_index = s3.idxmin()
s3_max_index = s3.idxmax()
print("Index value of the minimum element of s3:", s3_min_index)
print("Index value of the maximum element of s3:", s3_max_index)
```

Using Iris data, plot the following with proper legend and axis labels: (Download IRIS data from: https://archive.ics.uci.edu/ml/datasets/iris or import it from sklearn datasets)

a. Load data into pandas’ data frame. Use pandas.info () method to look at the info on datatypes in the dataset.

b. Find the number of missing values in each column (Check number of null values in a column using df.isnull().sum())

c. Plot bar chart to show the frequency of each class label in the data.

d. Draw a scatter plot for Petal Length vs Sepal Length and fit a regression line

e. Plot density distribution for feature Petal width.

f. Use a pair plot to show pairwise bivariate distribution in the Iris Dataset.

g. Draw heatmap for any two numeric attributes

h. Compute mean, mode, median, standard deviation, confidence interval and standard error for each numeric feature

i. Compute correlation coefficients between each pair of features and plot heatmap

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from scipy import stats

# Load the Iris dataset
iris_data = load_iris()
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris_df['target'] = iris_data.target
```

```python
# Task a: Bar chart of class label frequencies
class_counts = iris_df['target'].value_counts()
class_labels = iris_data.target_names
plt.bar(class_labels, class_counts)
plt.xlabel('Class Labels')
plt.ylabel('Frequency')
plt.title('Frequency of Each Class Label')
plt.show()
```

```python
# Task b: Scatter plot of Petal width vs Sepal width with regression line
sns.regplot(x='sepal width (cm)', y='petal width (cm)', data=iris_df)
plt.xlabel('Sepal width (cm)')
plt.ylabel('Petal width (cm)')
plt.title('Petal width vs Sepal width')
plt.show()
```

```python
# Task c: Density distribution plot of Petal length
sns.kdeplot(data=iris_df['petal length (cm)'], fill=True)
plt.xlabel('Petal length (cm)')
plt.ylabel('Density')
plt.title('Density Distribution of Petal length')
plt.show()
```

```python
# Task d: Pair plot for pairwise bivariate distribution
sns.pairplot(iris_df, hue='target')
plt.show()
```

```python
# Task e: Heatmap of correlation between numeric attributes
numeric_attributes = iris_df.drop('target', axis=1)
correlation_matrix = numeric_attributes.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
statistics = iris_df.describe()
confidence_interval = stats.t.interval(0.95,
                                       len(iris_df) - 1,
                                       loc=statistics.loc['mean'],
                                       scale=stats.sem(iris_df))
conf_interval_df = pd.DataFrame(confidence_interval, index=['lower_bound', 'upper_bound'])
statistics = pd.concat([statistics, conf_interval_df])
print(statistics)
```

```python
# Task g: Compute correlation coefficients and plot heatmap
correlation_matrix = iris_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Using Titanic dataset, to do the following:

a. Clean the data by dropping the column which has the largest number of missing values.

b. Find total number of passengers with age more than 30

c. Find total fare paid by passengers of second class

d. Compare number of survivors of each passenger class

e. Compute descriptive statistics for age attribute gender wise

f. Draw a scatter plot for passenger fare paid by Female and Male passengers separately

g. Compare density distribution for features age and passenger fare

h. Draw the pie chart for three groups labelled as class 1, class 2, class 3 respectively displayed in different colours. The occurrence of each group converted into percentage should be displayed in the pie chart. 
Appropriately Label the chart.

i. Find % of survived passengers for each class and answer the question “Did class play a role in survival?”.

```python
# Read the dataset from the Excel file
df = pd.read_excel("titanic3.xls")
```

```python
# a. Clean the data by dropping the column which has the largest number of missing values.
# Find the column with the largest number of missing values
missing = df.isnull().sum()
max_missing = missing.max()
col = missing[missing == max_missing].index[0]
print(f"The column with the largest number of missing values is {col}.")

# Drop the column
df = df.drop(col, axis=1)
print("The column has been dropped.")
```

```python
# b. Find total number of passengers with age more than 30
# Filter the dataframe by age
age_filter = df["age"] > 30
df_age = df[age_filter]

# Count the number of passengers
num_passengers = df_age.shape[0]
print(f"The total number of passengers with age more than 30 is {num_passengers}.")
```

```python
# c. Find total fare paid by passengers of second class
# Filter the dataframe by class
class_filter = df["pclass"] == 2
df_class = df[class_filter]

# Sum the fare column
total_fare = df_class["fare"].sum()
print(f"The total fare paid by passengers of second class is {total_fare:.2f}.")
```

```python
# d. Compare number of survivors of each passenger class
# Group the dataframe by class and survival
df_group = df.groupby(["pclass", "survived"])

# Count the number of passengers in each group
df_count = df_group.size().unstack()

# Print the result
print("The number of survivors of each passenger class is:")
print(df_count)
```

```python
# e. Compute descriptive statistics for age attribute gender wise
# Group the dataframe by gender
df_gender = df.groupby("sex")

# Compute the descriptive statistics for age
df_stats = df_gender["age"].describe()

# Print the result
print("The descriptive statistics for age attribute gender wise are:")
print(df_stats)
```

```python
# f. Draw a scatter plot for passenger fare paid by Female and Male passengers separately
# Import matplotlib library
import matplotlib.pyplot as plt

# Create a figure and two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot the fare paid by female passengers
ax1.scatter(df[df["sex"] == "female"]["age"], df[df["sex"] == "female"]["fare"], color="pink")
ax1.set_xlabel("Age")
ax1.set_ylabel("Fare")
ax1.set_title("Fare paid by female passengers")

# Plot the fare paid by male passengers
ax2.scatter(df[df["sex"] == "male"]["age"], df[df["sex"] == "male"]["fare"], color="blue")
ax2.set_xlabel("Age")
ax2.set_ylabel("Fare")
ax2.set_title("Fare paid by male passengers")

# Show the plot
plt.show()
```

```python
# g. Compare density distribution for features age and passenger fare
# Import seaborn library
import seaborn as sns

# Create a figure and two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot the density distribution for age
sns.kdeplot(df["age"], ax=ax1, color="green")
ax1.set_xlabel("Age")
ax1.set_title("Density distribution for age")

# Plot the density distribution for fare
sns.kdeplot(df["fare"], ax=ax2, color="orange")
ax2.set_xlabel("Fare")
ax2.set_title("Density distribution for fare")

# Show the plot
plt.show()
```

```python

# h. Draw the pie chart for three groups labelled as class 1, class 2, class 3 respectively displayed in different colours. The occurrence of each group converted into percentage should be displayed in the pie chart. 
# Appropriately Label the chart.
# Count the number of passengers in each class
class_count = df["pclass"].value_counts()

# Convert the count to percentage
class_percent = class_count / class_count.sum() * 100

# Define the labels and colors
labels = ["Class 1", "Class 2", "Class 3"]
colors = ["red", "yellow", "green"]

# Create a figure and a subplot
fig, ax = plt.subplots(figsize=(5, 5))

# Plot the pie chart
ax.pie(class_percent, labels=labels, colors=colors, autopct="%1.1f%%")
ax.set_title("Percentage of passengers by class")

# Show the plot
plt.show()
```

```python
# i. Find % of survived passengers for each class and answer the question “Did class play a role in survival?”.
# Compute the percentage of survived passengers for each class
survival_percent = df_count[1] / df_count.sum(axis=1) * 100

# Print the result
print("The percentage of survived passengers for each class is:")
print(survival_percent)

# Answer the question
answer = "Yes, class played a role in survival. The higher the class, the higher the survival rate."
print(answer)
```

FamilyName  Gender      MonthlyIncome (Rs.)

Shah        Male        44000.00

Vats        Male        65000.00

Vats        Female      43150.00

Kumar       Female      66500.00

Vats        Female      255000.00

Kumar       Male        103000.00

Shah        Male        55000.00

Shah        Female      112400.00

Kumar       Female      81030.00

Vats        Male        71900.00

Write a program in Python using Pandas to perform the following:

a. Calculate and display familywise gross monthly income.

b. Display the highest and lowest monthly income for each family name

c. Calculate and display monthly income of all members earning income less than Rs. 80000.00.

d. Display total number of females along with their average monthly income 

e. Delete rows with Monthly income less than the average income of all members


```python
# Create a data frame from the given data
data = {"FamilyName": ["Shah", "Vats", "Vats", "Kumar", "Vats", "Kumar", "Shah", "Shah", "Kumar", "Vats"],
        "Gender": ["Male", "Male", "Female", "Female", "Female", "Male", "Male", "Female", "Female", "Male"],
        "MonthlyIncome": [44000.00, 65000.00, 43150.00, 66500.00, 255000.00, 103000.00, 55000.00, 112400.00, 81030.00, 71900.00]}
df = pd.DataFrame(data)
```

```python
# a. Calculate and display familywise gross monthly income
familywise_income = df.groupby("FamilyName")["MonthlyIncome"].sum()
print("Familywise gross monthly income:")
print(familywise_income)
```

```python
# b. Display the highest and lowest monthly income for each family name
highest_income = df.groupby("FamilyName")["MonthlyIncome"].max()
lowest_income = df.groupby("FamilyName")["MonthlyIncome"].min()
print("Highest monthly income for each family name:")
print(highest_income)
print("Lowest monthly income for each family name:")
print(lowest_income)
```

```python
# c. Calculate and display monthly income of all members earning income less than Rs. 80000.00
low_income = df[df["MonthlyIncome"] < 80000.00]["MonthlyIncome"]
print("Monthly income of all members earning income less than Rs. 80000.00:")
print(low_income)
```

```python
# d. Display total number of females along with their average monthly income
female_count = df[df["Gender"] == "Female"].shape[0]
female_average = df[df["Gender"] == "Female"]["MonthlyIncome"].mean()
print("Total number of females:", female_count)
print("Average monthly income of females:", female_average)
```

```python
# e. Delete rows with Monthly income less than the average income of all members
average_income = df["MonthlyIncome"].mean()
df = df[df["MonthlyIncome"] >= average_income]
print("Data frame after deleting rows with Monthly income less than the average income of all members:")
print(df)
```
